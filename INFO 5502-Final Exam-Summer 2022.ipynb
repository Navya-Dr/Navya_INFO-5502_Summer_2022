<span style='background:yellow'>**Question 2 (100 pt).**</span> Write a python program to evaluate the text classification performance using accuracy, precision, recall, and f1 score. 

Accuracy = (TP + TN) / (TP+ TN+ FP+ FN)

Recall = (TP) / (TP+FN)

Precision = (TP) / (TP+FP)

F1 Score = (2*Precision*Recall) / (Precision + Recall)

Here is the confusion matrix of a text classfication task (six categories) on the testing data. 



```python
#### You code here
import numpy as np

confusion_matrix = [[157,46,25,22,15,3],
                    [28,151,45,18,1,6],
                    [11,38,135,14,3,7],
                    [11,15,18,70,6,4],
                    [14,12,14,7,56,2],
                    [12,7,6,5,5,16]]
conf_arr = np.array(confusion_matrix)

for i in range(len(conf_arr)):
  acc,recall,precision,f1_score = 0,0,0,0
  tp = conf_arr[i,i]
  fp = conf_arr[i,:].sum() - tp
  fn = conf_arr[:,i].sum() - tp
  tn = conf_arr.sum().sum() - (tp+fp+fn)
    
  acc= (tp + tn) / conf_arr.sum().sum()
  recall = tp / (tp + fn)
  precision = tp / (tp + fp)
  f1_score = 2 * precision * recall / (precision + recall)
    
  print('*************************************************')
  print(' Text classification Perfomance for Class ',(i+1))
  print('Accuracy : ',acc)
  print('Recall : ',recall)
  print('Precision: ',precision)
  print('F1 Score: ',f1_score)
  print()




```

```python
#### You code here


#1.

intru_data = pd.read_csv('https://raw.githubusercontent.com/unt-iialab/info5502-spring2022/main/finalexam/q4/inetd-intrusion.int', sep = " ", header = None)
nor_data = pd.read_csv('https://raw.githubusercontent.com/unt-iialab/info5502-spring2022/main/finalexam/q4/inetd-normal.int', sep = " ", header = None)

intru_data.columns = ['PID','system_call']
print(intru_data.head())
print("___________________________________________")
nor_data.columns = ['PID','system_call']
print(nor_data.head())

intru_dict = {}
nor_dict = {}

for pid in list(set(intru_data['PID'].values)):
    intru_dict[pid] = " ".join(str(i) for i in list(intru_data[intru_data['PID'] == pid]['system_call'].values))

for pid in list(set(nor_data['PID'].values)):
    nor_dict[pid] = " ".join(str(j) for j in list(nor_data[nor_data['PID'] == pid]['system_call'].values))
```

```python
#2.
import re
from nltk.util import ngrams
intru_list = []
nor_list = []

for key in intru_dict.keys():
    k_val = intru_dict[key]
    tks = [i for i in k_val.split(" ") if i != ""]
    
    for i in list(ngrams(tks, 25)):
        intru_list.append(i)
        
for j in nor_dict.keys():
    k_vl = nor_dict[j]
    tks = [i for i in k_vl.split(" ") if i != ""]
    
    for i in list(ngrams(tks, 25)):
        nor_list.append(i) 

print("For Intrusion List: ",len(intru_list))
print("For Normal List: ",len(nor_list))
```

```python
#3.
from sklearn.tree import DecisionTreeClassifier
from sklearn import preprocessing
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split

intru_df = pd.DataFrame(intru_list)
nor_df = pd.DataFrame(nor_list)

intru_df['class'] = 1
nor_df['class'] = 0

data_fr = pd.DataFrame()

data_fr = data_fr.append(intru_df)
data_fr = data_fr.append(nor_df)
```

```python
#4.

data_fr = data_fr.drop_duplicates(subset = data_fr.columns[:-1], keep = 'last')
data_fr['class'].value_counts()

data_fr.columns[:-1]
```

```python
#5.

x = data_fr.drop("class", axis = 1)
y = data_fr['class']
```

```python
#6.
x_train, x_valid, y_train, y_valid = train_test_split(x, y, test_size = 0.3, random_state = 0)

classifier = DecisionTreeClassifier(max_depth = 20)
classifier.fit(x_train, y_train)

y_pred = classifier.predict(x_valid)

print(classification_report(y_valid, y_pred))
```

The impact of customers’ reviews on customer’s decision making is significant. Therefore, businesses try to increase this impact by increasing the ratio of the positive reviews to negative reviews. One effective way to increase the ratio is offering the incentives such as small gifts or free next services to customers to post the reviews for item they purchase or used. However, it is not always the case, and customer may post their reviews without receiving or accepting the incentives. So, the reviews could be in either incentivized or non-incentivized category.


Considering the importance of reviews, in this question you need to compare these two categories of reviews. One method to compare the effectiveness of two categories is conducting the A/B testing.

(2)(50 points) Apply an A/B test": consider the “incentivized” column as main column for A & B groups.

a. The “incentivized” column includes ‘NominalGift’, ‘VendorReferredIncentivized’, ‘NoIncentive’, ‘NonNominalGift’, ‘VendorReferred’, ‘None’

b. Group A (Incentive): ‘NominalGift’, and ‘VendorReferredIncentivized’

c. Group B (NoIncentive): ‘NoIncentive’, ‘NonNominalGift’, ‘VendorReferred’

d. Make the decision if ‘None’ includes in either Group A or B or remove that.

e. The “OverAllRating” is another column that is used to compare groups A and B.

Which one A or B group has higher “OverAllRating”?

```python
#### You code and your answer here here

import pandas as pd
from scipy.stats import norm
import matplotlib.pyplot as plt

data_fr = pd.read_csv("https://raw.githubusercontent.com/unt-iialab/info5502-spring2022/main/finalexam/Software_Reviews_csv_files_S.csv")
data_fr.head()

```

```python
print(data_fr.info())
print("_____________________________________________________________________________________________")
print(data_fr.columns)

```

```python
data_fr.drop(['Unnamed: 0', 'Unnamed: 0.1','Unnamed: 27'], axis=1)
```

```python
Group_A = data_fr[(data_fr['incentivized'] == "NominalGift") | (data_fr['incentivized'] == "VendorReferredIncentivized")]['OverAllRating']
Group_B = data_fr[(data_fr['incentivized'] == "NoIncentive") | (data_fr['incentivized'] == "NonNominalGift") | (data_fr['incentivized'] == "VendorReferred")]['OverAllRating']
```

```python
no_of_ctrl_grp = Group_A.shape[0]
no_of_trt_grp = Group_B.shape[0]

alpa = 0.05

mean_ctrl_grp = np.mean(Group_A)
mean_trt_grp = np.mean(Group_B)

Group_A = Group_A.astype('float')
Group_B = Group_B.astype('float')

var_ctrl_grp = np.var(Group_A)
var_trt_grp = np.var(Group_B)

comb_var = np.sqrt(var_ctrl_grp / no_of_ctrl_grp + var_trt_grp / no_of_trt_grp)

t_test = (mean_ctrl_grp - mean_trt_grp) / np.sqrt(var_ctrl_grp / no_of_ctrl_grp + var_trt_grp / no_of_trt_grp)

p_value = norm.sf(t_test) * 2

z_critical = norm.ppf(1-alpa/2)

mar_error = z_critical * comb_var

Conf_Interval = [(mean_ctrl_grp - mean_trt_grp) - mar_error, (mean_ctrl_grp - mean_trt_grp) + mar_error]

print("Test Stats: ", t_test)
print("Z-critical Values: ", z_critical)
print("P_value: ", p_value)
print("Confidence Interval of Z-test for proportions: ", np.round(Conf_Interval,2))
```

```python
z = np.arange(-3 ,3, 0.1 )
plt.plot(z, norm.pdf(z), label = 'Std Normal Distribution',color = 'blue',linewidth = 1.5)
plt.fill_between(z[z > z_critical], norm.pdf(z[z > z_critical]), label = 'Right Rejection Region',color ='r' )
plt.fill_between(z[z < (-1) * z_critical], norm.pdf(z[z < (-1) * z_critical]), label = 'Left Rejection Region',color ='g' )
plt.title('Rejection Region')
plt.legend()
plt.show()
```
