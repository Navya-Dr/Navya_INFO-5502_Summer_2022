```python
# Your answer here (python code + explanation):
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')

url='https://raw.githubusercontent.com/suthapalliuday/INFO5502-Summer2022/main/datasets/lab_assignment_05_census.csv'
df=pd.read_csv(url)
print(df)
#General bar graph showcasing view of data
crosstb = pd.crosstab(df.Sex, df.Age)
barplot = crosstb.plot.bar(rot=0)

#Group by Sex and year
counts = df.groupby(['Sex','Year']).Age.count().unstack()
print(counts)
counts.plot(kind='bar', stacked=True)

#Group by all Sex,Age and Year
counts2 = df.groupby(['Sex','Age','Year']).Age.count().unstack()
counts2.plot(kind='bar', stacked=True)


#Comparing co-occurences
ax = counts.plot(kind='bar',stacked = False, colormap = 'Paired')

for p in ax.patches:
        ax.annotate(np.round(p.get_height(),decimals=0).astype(np.int64), (p.get_x()+p.get_width()/2., p.get_height()), ha='center', va='center', xytext=(2, 10), textcoords='offset points')

plt.xlabel ('Sex')
plt.ylabel ('Co-Occurences ')
plt.title('Comparison Of Occurences  In An Age Group',fontsize=10)
plt.show()

'''
The code showcases different ways of visualizing data using barplots in different scenarios of filtering.Used standard Matplot libarary ,
pandas and numpy to plot and read data.
'''
```
